```python
import pandas as pd

zip_codes_df = pd.read_csv(r"C:\Users\nrdee\Documents\Data\Data Science Masters\Data 780\Final-Project-DATA780\data_sources\us_zipcodes\zip_code_database.csv")
# counties_filter = ['Alameda County','San Francisco County','Santa Clara County','San Mateo County','Los Angeles County','San Diego County']
# zip_codes_df2 = zip_codes_df[zip_codes_df['county'].isin(counties_filter)].reset_index(drop=True)

zip_codes_df2 = zip_codes_df[['zip','primary_city','county','latitude','longitude','irs_estimated_population']]
zip_codes_df2
```

```python
geo_corr_df = pd.read_csv(r"C:\Users\nrdee\Documents\Data\Data Science Masters\Data 780\Final-Project-DATA780\data_sources\us_zipcodes\geocorr2022_2518309437.csv")
geo_corr_df = geo_corr_df[1:].reset_index(drop=True)
geo_corr_df
```

#### Merge our geo_corr data and zip code data!

```python
# Ensure column types match
zip_codes_df2['zip'] = zip_codes_df2['zip'].astype(str)
geo_corr_df['zcta'] = geo_corr_df['zcta'].astype(str)

#inner join the zips and geo_corr
merged_df = pd.merge(zip_codes_df2, geo_corr_df, left_on='zip', right_on='zcta', how='inner')
merged_df2 = merged_df[['zip','primary_city','county','latitude','longitude','puma22','PUMA22name','irs_estimated_population','hus20']]
merged_df3 = merged_df2.rename(columns={'irs_estimated_population': 'zip_estimated_pop', 'hus20': 'housing_units_2020'})
merged_df3
```

```python
# render data down so that the top zip code in terms of housing units is picked, drop duplicates
top_zip_per_puma = merged_df3.sort_values('housing_units_2020', ascending=False).drop_duplicates('puma22')
top_zip_per_puma = top_zip_per_puma[['puma22', 'zip', 'primary_city', 'county', 'latitude', 'longitude']].reset_index(drop=True)
top_zip_per_puma
```
