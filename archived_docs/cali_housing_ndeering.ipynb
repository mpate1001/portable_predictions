```python
import kagglehub
import pandas as pd
from pathlib import Path
import os
import geopandas
import matplotlib.pyplot as plt
from shapely.geometry import Point

'''
This will prompt for kaggle.json path or use default ~/.kaggle/kaggle.json.
For me this is saved locally on my machine. You will need to setup your own api key at Kaggle to grab the code:
https://www.kaggle.com/docs/api#interacting-with-datasets
'''
kagglehub.login(r"C:\Users\nrdee\Documents\Data\Data Science Masters\Data 780\kaggle.json")
path = kagglehub.dataset_download("camnugent/california-housing-prices")
print("Path to dataset files:", path)

csv_path = os.path.join(path, "housing.csv")
cali_df = pd.read_csv(csv_path)
cali_df.head(20)
```

```python
import folium
import branca.colormap as cm
from folium.plugins import MarkerCluster

cali_df = cali_df.dropna(subset=["latitude", "longitude", "median_house_value"])

#set up color scale
min_val = cali_df["median_house_value"].min()
max_val = cali_df["median_house_value"].max()

#color filter, find more info here: https://python-visualization.github.io/folium/latest/advanced_guide/colormaps.html
colormap = cm.linear.YlOrRd_09.scale(min_val, max_val)

#create map
m = folium.Map(location=[36.7783, -119.4179], zoom_start=6)

#add colored circles
for _, row in cali_df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=4,
        color=colormap(row["median_house_value"]),
        fill=True,
        fill_opacity=0.7,
        popup=f"${row['median_house_value']:,.0f}"
    ).add_to(m)

#add color legend to map
colormap.caption = "Median House Value"
colormap.add_to(m)

#save map
m.save("california_colored_prices.html")
```
